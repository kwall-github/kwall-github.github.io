```python
import time
print("Run time: {0}".format(time.strftime("%Y-%m-%d %H:%M")))
import pandas as pd
%matplotlib inline
import cb_utils
cb_utils.info()
hitting_obj = cb_utils.hitting()
```

# 1 Plot of all hitters AB/OPS by ballot position

All players with hitting statistics are classified into a ballot position which can be one of C, 1B, 2B, 3B, SS, OF, DH.


For players in NAIA, NCAA, NJCAA, ballot position was determined by looking at the box scores. Each players season AB were broken down by position and the dominant position was selected.

Boxscore were not collected for CCCAA, NWAC and USCAA. Some of these orgs had ballot worthy players and their ballot position is dettermined by their team roster position.  The players who did not have enough AB or had low OPS were not assigned a ballot position.

## Action required: eleven players not assigned ballot position
Eleven ballot worthy players are not classified into a ballot position. 

* Nine players in NAIA, NCAA, NJCAA players have their season AB split evenly among different positions (no dominant).   
* One player in CCCAA and one player in NWAC, have vague roster position (either INF or 1B/OF).  Will need more information to classify.

These players will need to be classified manually or we might consider a new ballot position for these 'multi-position' players.




```python
hitting_obj = cb_utils.hitting()
hitting_obj.plot_players()
```

# 2 Hitting ballot positions

Below there is a section for each ballot position, which shows a plot of all players assigned to the ballot position, followed by a table with the selected players and their season stats.

Basic selection criteria: AB >= 100 and OPS >= 0.700

Each plot has two blue division lines that represent the AB and OPS thresholds for that position.  Only players in the UPPER RIGHT area will be selected.

We can adjust the AB and OPS threshold for each position so we can get the desired number of players on the ballot for that position.  

```python
pos_df_d = dict()
```

## Catchers (C)

```python
pos = 'C'
pos_df_d[pos] = hitting_obj.select(positions=[pos], ab_min=100, ops_min=0.700, plot=True)
pos_df_d[pos]
```

## Players without dominant position (Need help classifying)

```python
pos = '***No dominant position'
pos_df_d[pos] = hitting_obj.select(positions=[pos], ab_min=100, ops_min=0.700, plot=True)
pos_df_d[pos]
```

## Build and write the ballot csv file

```python
dfs = [pos_df_d[pos] for pos in ('C', '1B', '2B', '3B', 'SS', 'OF', 'DH', '***No dominant position')]
ballot_df = pd.concat(dfs)
fname = r"2017 CBN Hitting Ballot.csv"
print("...Writing to {0}".format(fname))
ballot_df.to_csv(fname)
```
