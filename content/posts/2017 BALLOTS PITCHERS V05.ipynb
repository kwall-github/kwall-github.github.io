```python
import time
print("Run time: {0}".format(time.strftime("%Y-%m-%d %H:%M")))
import pandas as pd
%matplotlib inline
import cb_utils
cb_utils.info()
pitching_obj = cb_utils.pitching()
```

# 1 Plot of all pitchers IP/ERA by ballot position

All players with pitching statistics are classified into a ballot position which can be one of LHP, RHP.

Pitcher handedness was determined by the CBN CIC web page.

There are a small number of players with pitching stats whose handedness is not shown on the CBN CIC web page (Those players are shown in red in the plot below).  Since the pitching stats for these players did not qualify for ballot selection no effort was made to classify as LHP or RHP. 


```python
pitching_obj.plot_players()
```

# 2 Pitching ballot positions

Below there is a section for each ballot position, which shows a plot of all players assigned to the ballot position, followed by a table with the selected players and their season stats.

Basic selection criteria: IP >= 60 and ERA <= 5.0

Each plot has two blue division lines that represent the IP and ERA thresholds for that position.  Only players in the LOWER RIGHT area will be selected.

We can adjust the IP and ERA threshold for each position so we can get the desired number of players on the ballot for that position.  

```python
pos_df_d = dict()
```

## Left Handed Pitchers (LHP)

```python
pos = 'LHP'
pos_df_d[pos] = pitching_obj.select(positions=[pos], ip_min=60, era_max=5.0, plot=True)
pos_df_d[pos]
```

## Right Handed Pitchers (RHP)

```python
pos = 'RHP'
pos_df_d[pos] = pitching_obj.select(positions=[pos], ip_min=60, era_max=5.01, plot=True)
pos_df_d[pos]
```

## Unknown Handed Pitchers (***)

```python
pos = '***'
pos_df_d[pos] = pitching_obj.select(positions=[pos], ip_min=50, era_max=5.0, plot=True)
pos_df_d[pos]
```

## Closers (Closer)
The criteria for selecting closers is:

SV >= 2  this might be too harsh for closers on weak teams that never have the lead

GS < 5  allow some players to start a few games, but too many GS eliminates as a pure closer (too harsh?)

ERA <= 5.0  


```python
pos = 'Closer'
pos_df_d[pos] = pitching_obj.select_saves(save_min=2, era_max=5.0, plot=False)
pos_df_d[pos]
```

## Build and write the ballot csv file

```python
dfs = [pos_df_d[pos] for pos in ('LHP', 'RHP', 'Closer')]
ballot_df = pd.concat(dfs)
fname = r"2017 CBN Pitching Ballot.csv"
print("...Writing to {0}".format(fname))
ballot_df.to_csv(fname)
```
